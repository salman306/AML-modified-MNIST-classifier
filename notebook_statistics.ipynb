```python
import pandas as pd
import numpy as np

X = pd.read_csv("train_x.csv", header = None)
Y = pd.read_csv("train_y.csv", header = None)
```

```python
X_COPY = X
Y_COPY = Y
```

```python
#merge
data = X_COPY
data['Label'] = Y_COPY
```

```python
# get categorical group
data_len = len(data.index)
group_data = pd.DataFrame(data.groupby('Label').size().apply(lambda x : x * 100.0 / data_len))
group_data.columns = ['Percentage']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,5))
g = sns.barplot(x=group_data.index, y=group_data.Percentage)
count = 0
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
           height + 0.26,
           str(round(group_data.iloc[count, 0],1)) + "%",
           ha="center")
    count+=1
plt.show()
```
